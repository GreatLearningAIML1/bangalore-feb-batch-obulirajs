### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
iqr_age_75 = hrdata['Age'].quantile(.75)
iqr_age_25 = hrdata['Age'].quantile(.25)
print('IQR 75 :' + str(iqr_age_75))
print('IQR 25 :' + str(iqr_age_25))
iqr_age = iqr_age_75 - iqr_age_25
print("IQR of Age :" + str(iqr_age))
```
