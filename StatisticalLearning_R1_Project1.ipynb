```python
import pandas as pd
import numpy as np
```

```python
# 1- Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.
df = pd.read_csv("Laliga.csv")
# df.describe()
df = df.replace('-', 0)
df.head(10)
```

```python
# 2 - Print all the teams which have started playing between 1930-1980

# Function to get the Debut year.
# If the value doesnot fall between years 1930-1980 return Zero
def get_DebutYear(debut):
#     from_year = 1929
#     to_year = 1980
    result = ''
    result = debut.split('-')
    if len(result) < 2 :
        return result[0]
    else:
        result = int(result[0]) #int(debut.split('-')[0])
        if (result > 1929 and result <= 1980):
            return result
        else:
            return 0

# Create a new Column nd append to existing dataframe
df['DebutYear'] = df['Debut'].apply(get_DebutYear)

# Convert data to integer type
df['DebutYear'] = df['DebutYear'].astype(np.int64)

#Filter the  data which are greater than Zero
result = df[['Team','Debut']][(df['DebutYear'] >= 1930) ].sort_values(by='Debut', ascending=True)
result.reset_index(drop=True)
# result.shape
```

```python
#3 - Print the list of teams which came Top 5 in terms of points (2.5 points)

df['Points'] = df['Points'].astype(np.int64)
df.sort_values(by='Points', ascending=False)
df.nlargest(5,'Points')
```

```python
#Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
#Using the same function, find the team which has maximum and minimum goal difference. (5 points)

def Goal_diff_count(): #GoalsFor, GoalsAgainst):
    #Convert datatype of the columns to integer
    df['GoalsFor'] = df['GoalsFor'].astype('int64')
    df['GoalsAgainst'] = df['GoalsAgainst'].astype('int64')
    #Calculate the goal difference 
    df['Goal_diff_count'] = df['GoalsFor'] - df['GoalsAgainst']
    #Find the maximum value 
    max_index = df['Goal_diff_count'].idxmax()
    print('Team with Max Goal Difference is -->' + str(df.iloc[[max_index]]['Team'].values[0]))
    min_index = df['Goal_diff_count'].idxmin()
    #Find minimmum value
    print('Team with Minimum Goal Difference is --> ' , str(df.iloc[[min_index]]['Team'].values[0]))
    return df['Goal_diff_count']

goal_diff = Goal_diff_count()
# goal_diff
# df.head()

```

```python
# Create a new column with name “Winning Percent” and append it to the data set (5 points)
# Percentage of Winning = (GamesWon / GamesPlayed)*100

df['GamesWon'] = df['GamesWon'].astype('int64') 
df['GamesPlayed'] = df['GamesPlayed'].astype('int64')
df['Winning Percent'] = (df['GamesWon'] / df['GamesPlayed']) * 100
df['Winning Percent'].replace("",0)
# df['Winning Percent'].head()
# If there are any numerical error, replace it with 0%

# Print the top 5 teams which has the highest Winning percentage
df.sort_values(by='Winning Percent', ascending=False).nlargest(5, 'Winning Percent')
```

```python
# Group teams based on their “Best position” and print the sum of their points for all positions (5 points)
# Eg: Best Position   Points
#        1            25000
#        2            7000

df.groupby(['BestPosition'])[['Points']].agg('sum')
```
